MATPLOTLIB

```python
import matplotlib.pyplot as plt  #import the pyplot module from MATPLOTLIB package
import numpy as np
x = np.linspace(0,2,10)  #creates the x axis
y1 = 3*x+2+np.random.rand(10)  #This is first data and random adds noise to the data
y2 = 3*x+2
plt.figure(figsize = (4,3),dpi = 300)
plt.scatter(x,y1,c = "red",label = "Plot with Noise")  #this creates a red colored scatter plot
plt.plot(x,y2,c = "black",linestyle = '--',linewidth = 2, marker = '*', markersize = 8,label = "Actual plot") #This creates a line plot
plt.xlabel("x") #This creates a x-label
plt.ylabel("y") #This creates a y-label
plt.legend()
plt.title("My first plot in python", fontdict = {'fontsize' : 18})  #Adds the title

```

Creating subplots using MATPLOTLIB

```python
plt.figure(figsize = (3,6))  #Create a new figure
plt.subplot(2,1,1)  #This creates a subplot located at position 1 in a 2x1 matrix
plt.scatter(x,y1,c = "red")  #this creates a red colored scatter
plt.xlabel("x") #This creates a x-label
plt.ylabel("y") #This creates a y-label
plt.title("Plot with noise")

plt.subplot(2,1,2)  #This creates a subplot located at position 2 in a 2x1 matrix
plt.plot(x,y2,c = "black")  #this creates a red colored scatter
plt.xlabel("x") #This creates a x-label
plt.ylabel("y") #This creates a y-label
plt.title("Actual Plot")

plt.show()  #This displays the figure in the console
```
